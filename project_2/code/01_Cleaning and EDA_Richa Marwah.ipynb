## Reading in Dataset

```python
#Reading in training dataset
ames_train = pd.read_csv('../datasets/train.csv')
ames_train.head()
```

```python
#Reading in test dataset
ames_test = pd.read_csv('../datasets/test.csv')
ames_test.head()
```

## Data Cleaning

```python
ames_train.shape
```

<i> The training dataset has 81 variables and 2051 observations (each representing a sale). As each sale has a unique Id, we can set the Id column as our index. </i>

```python
ames_train = ames_train.set_index('Id', drop=True)
ames_train.head()
```

```python
ames_test.shape
```

<i> The test dataset has 80 variables and 879 observations (each representing a sale). As each sale has a unique Id, we can set the Id column as our index. </i>

```python
ames_test = ames_test.set_index('Id', drop=True)
ames_test.head()
```

<i> We can drop the PID column, as it represents parcel identification number and is not likely to influence our target variable, SalePrice. </i>

```python
ames_train.drop('PID', axis = 1, inplace=True)
ames_test.drop('PID', axis = 1, inplace=True)
```

```python
#Inspecting the remaining data types
ames_train.info()
```

```python
#Inspecting the remaining data types
ames_test.info()
```

<i> Upon examination of the data types, we notice that the MS SubClass is represented as an int value even though it is a categorical variable. We will transform this to a string for ease of further analysis.</i>

```python
ames_train['MS SubClass'] = ames_train['MS SubClass'].astype(str)
ames_test['MS SubClass'] = ames_test['MS SubClass'].astype(str)
```

<i> While Overall Qual and Overall Cond are categorical variables too, we will model them as continuous features for ease of analysis and let them remain as int values at this time. </i>

### Determining what missing values mean

<i> We observe that many columns have null values. Upon consulting the provided data dictionary, we conclude that the null values are not missing values but rather show that the particular feature is not available in a given property. For instance, there are 1911 null values under the Alley column, showcasing that 1911 houses do not have alley access. </i>

<i> However, as majority of the properties do not have Alley and Misc Features (more than 95% of the values in these columns are null), we will drop these columns. In addition, we will drop Misc Val as it is associated with the value of miscellaneous features. Similarly, the null values in the Pool QC column suggest that most properties do not have a pool. As such, we will also drop the Pool Area and Pool QC columns. </i>

```python
ames_train = ames_train.drop(['Alley', 'Pool Area','Pool QC', 'Misc Feature', 'Misc Val'], axis = 1)
ames_test = ames_test.drop(['Alley', 'Pool Area','Pool QC', 'Misc Feature', 'Misc Val'], axis = 1)
```

```python
ames_train.shape
```

```python
ames_test.shape
```

### Imputing null values

<i> Next, we observe that Fence and Fireplace Qu columns have a significant number of null values. We will replace the null values with zero, to represent absence of the feature. </i>

```python
ames_train['Fence'].fillna(0, inplace=True)
ames_train['Fireplace Qu'].fillna(0, inplace=True)
ames_test['Fence'].fillna(0, inplace=True)
ames_test['Fireplace Qu'].fillna(0, inplace=True)
```

<i> For the remaining null values in numerical data, we will replace with the column mean, where possible. Following which, we will drop the observations that have null values. </i>

```python
ames_train = ames_train.fillna(ames_train.mean())
ames_train.dropna(axis = 0, inplace=True) #We only drop values from the train set
ames_test = ames_test.fillna(ames_test.mean())
```

## Exploratory Data Analysis (EDA)

<i> First, we run a quick check on the spread of the target variable, SalePrice. <i>

```python
#Plotting a histogram to depict the distribution of SalePrice 
plt.hist(ames_train['SalePrice']);
plt.title('Distribution of SalePrice');
plt.axvline(ames_train['SalePrice'].mean(), color='k', linestyle='dashed', linewidth=2);
```

<i> SalePrice seems to be right-skewed with a majority of the sales falling below $400,000. Next, we examine the numeric variables against SalePrice.

```python
numerics = ['int64', 'float64']
```

```python
#Defining function to plot numeric variables against SalePrice
def subplot_scatter(df, list_of_columns):
    nrows = int(np.ceil(len(list_of_columns)/2)) 
    fig, ax = plt.subplots(nrows=nrows, ncols=2, figsize = (20, 100)) #fixing number of rows, cols and the figure size
    ax = ax.ravel() 
    for i, column in enumerate(list_of_columns): 
        ax[i].scatter(x = df[column] , y = ames_train['SalePrice'])
        ax[i].set_title(str(column) + ' against SalePrice') #plotting title
        ax[i].set_xlabel(column) #plotting x label
        ax[i].set_ylabel('SalePrice') #plotting y label
```

```python
ames_train_num = ames_train.select_dtypes(include=numerics).drop('SalePrice', axis = 1)
subplot_scatter(ames_train_num, ames_train_num.columns)
```

<i> We can see that variables such as Low Qual Fin SF and 3Ssn Porch have a very high number of zero values. We can drop them as they are not likely to have a significant influence on SalePrice. </i>  

```python
ames_train = ames_train.drop(['Low Qual Fin SF', '3Ssn Porch'], axis = 1)
ames_test = ames_test.drop(['Low Qual Fin SF', '3Ssn Porch' ], axis = 1)
```

<i> As the scatter plots above have shown, certain discrete variables such as Bsmt Full Bath, Bsmt Half Bath, Full Bath, Half Bath, Bedroom AbvGr, Kitchen AbvGr, TotRms AbvGrd, Fireplaces and Garage Cars are better represented as categorical as there are a small number of possible values that each of these variables can take. </i>

```python
ames_train['Bsmt Full Bath'] = ames_train['Bsmt Full Bath'].astype(str)
ames_train['Bsmt Half Bath'] = ames_train['Bsmt Half Bath'].astype(str)
ames_train['Full Bath'] = ames_train['Full Bath'].astype(str)
ames_train['Half Bath'] = ames_train['Half Bath'].astype(str)
ames_train['Bedroom AbvGr'] = ames_train['Bedroom AbvGr'].astype(str)
ames_train['Kitchen AbvGr'] = ames_train['Kitchen AbvGr'].astype(str)
ames_train['TotRms AbvGrd'] = ames_train['TotRms AbvGrd'].astype(str)
ames_train['Fireplaces'] = ames_train['Fireplaces'].astype(str)
ames_train['Garage Cars'] = ames_train['Garage Cars'].astype(str)
```

```python
ames_test['Bsmt Full Bath'] = ames_test['Bsmt Full Bath'].astype(str)
ames_test['Bsmt Half Bath'] = ames_test['Bsmt Half Bath'].astype(str)
ames_test['Full Bath'] = ames_test['Full Bath'].astype(str)
ames_test['Half Bath'] = ames_test['Half Bath'].astype(str)
ames_test['Bedroom AbvGr'] = ames_test['Bedroom AbvGr'].astype(str)
ames_test['Kitchen AbvGr'] = ames_test['Kitchen AbvGr'].astype(str)
ames_test['TotRms AbvGrd'] = ames_test['TotRms AbvGrd'].astype(str)
ames_test['Fireplaces'] = ames_test['Fireplaces'].astype(str)
ames_test['Garage Cars'] = ames_test['Garage Cars'].astype(str)
```

### Combining Features

<i> Using the Year Built and Yr Sold columns, we can calculate the age of the house when it is sold. </i>

```python
ames_train['Age Sold'] = ames_train['Yr Sold'].astype(int) - ames_train['Year Built'].astype(int)
ames_train.drop(['Yr Sold', 'Year Built'], axis = 1, inplace = True)
```

```python
ames_test['Age Sold'] = ames_test['Yr Sold'].astype(int) - ames_test['Year Built'].astype(int)
ames_test.drop(['Yr Sold', 'Year Built'], axis = 1, inplace = True)
```

<i> We can also combine the Bsmt Full Bath and Bsmt Half Bath to provide total basement bath count. </i>

```python
ames_train['TotBath Bsmt'] = ames_train['Bsmt Full Bath'].astype(float) + ames_train['Bsmt Half Bath'].astype(float)
ames_train.drop(['Bsmt Full Bath', 'Bsmt Half Bath'], axis = 1, inplace = True)
```

```python
ames_test['TotBath Bsmt'] = ames_test['Bsmt Full Bath'].astype(float) + ames_test['Bsmt Half Bath'].astype(float)
ames_test.drop(['Bsmt Full Bath', 'Bsmt Half Bath'], axis = 1, inplace = True)
```

<i> Likewise for the Full Bath and Half Bath columns. </i>

```python
ames_train['TotBath AbvGr'] = ames_train['Full Bath'].astype(int) + ames_train['Half Bath'].astype(int)
ames_train.drop(['Full Bath', 'Half Bath'], axis = 1, inplace = True)
ames_train.head()
```

```python
ames_test['TotBath AbvGr'] = ames_test['Full Bath'].astype(int) + ames_test['Half Bath'].astype(int)
ames_test.drop(['Full Bath', 'Half Bath'], axis = 1, inplace = True)
ames_test.head()
```

## Exploratory Visualizations

<i> Based on the histograms, a number of the variables appear to be right-skewed, such as BsmtFin SF 1, BsmtFin SF 2, Lot Area and Mas Vnr Area, to name a few. <i> 
We observe that variables such as Overall Cond, BsmtFin SF 2, Bsmt Unf SF, Enclosed Porch, Screen Porch and Mo Sold have very low correlation with SalePrice (all below +/- 0.2). As such, we will drop these variables ahead of further analysis. <br>   
We also see that the variables 1st Flr SF and Total Bsmt SF have a strong positive correlation (above 0.85), indicating multicollinearity. To address this, we will drop Total Bsmt SF as it is likely to be similar to the 1st Flr SF.
<i>
